```python
#veri çekme
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
import os

def scrape_youtube_comments(video_url, num_comments=50):
    driver = webdriver.Firefox() 
    driver.get(video_url)

    time.sleep(15)

    comments = []
    actions = ActionChains(driver)
    
    while len(comments) < num_comments:
        driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
        time.sleep(10)
        
        try:
            WebDriverWait(driver, 30).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "#content-text"))
            )
        except:
            print("Yorumlar yüklenemedi.")
            break

        comment_elements = driver.find_elements(By.CSS_SELECTOR, "#content-text")
        
        for element in comment_elements:
            comment = element.text.strip()
            if comment and comment not in comments:
                comments.append(comment)
            if len(comments) >= num_comments:
                break

    driver.quit()
    return comments


video_url = "https://www.youtube.com/watch?v=Nh7phB3QXk0"
comments = scrape_youtube_comments(video_url, num_comments=50)


csv_dosya_adi = "youtube.csv"


if os.path.exists(csv_dosya_adi):
    mevcut_df = pd.read_csv(csv_dosya_adi)
    yeni_df = pd.DataFrame(comments, columns=["comment"])
    sonuc_df = pd.concat([mevcut_df, yeni_df], ignore_index=True).drop_duplicates()
    sonuc_df.to_csv(csv_dosya_adi, index=False)
else:
    yeni_df = pd.DataFrame(comments, columns=["comment"])
    yeni_df.to_csv(csv_dosya_adi, index=False)

print(f"Yorumlar kaydedildi veya güncellendi: {csv_dosya_adi}")
```
